```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('extra_dep_data.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df
```

```python
df['File Name'] = df['File Name'].str.replace('.fastq', '', regex=False)
df = df.rename(columns={'File Name': 'Run ID'})
df['Condition'] = 'Depression'
new_order = ['Run ID', 'Condition'] + [col for col in df.columns if col not in ['Run ID', 'Condition']]
df = df[new_order]
```

```python
print(list(df.columns))
```

```python
df['Uncultured Organism'] += df[['Bacterium', 'Uncultured', 'Unidentified', 'Candidatus']].sum(axis=1)
df['Clostridium'] += df[['Clostridiales']].sum(axis=1)
df['Bacillus'] += df[['Bacillota']].sum(axis=1)
```

```python
df = df.drop(['Bacterium', 'Uncultured', 'Unidentified', 'Candidatus', 'Clostridiales', 'Bacillota', 
             'Bacteriophage', 'Caudoviricetes'], axis=1)
```

```python
numeric_cols = df.select_dtypes(include=['number'])

df[numeric_cols.columns] = numeric_cols.div(numeric_cols.sum(axis=1), axis=0)
```

```python
zero_counts = (df == 0.0).sum()
zero_counts
```

```python
threshold = 0.8 * len(df)
filtered = zero_counts[zero_counts <= threshold]
filtered
```

```python
filtered_df = df[filtered.index]
filtered_df = filtered_df.rename(columns={'File Name': 'Run ID'})
filtered_df.columns
```

```python
filtered_df
```

```python
filtered_df = filtered_df.fillna(0)
```

```python
filtered_df
```

```python
filtered_df = filtered_df[filtered_df.select_dtypes(include=[np.number]).sum(axis=1) != 0]
```
