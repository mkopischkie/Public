```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import normalized_mutual_info_score
from sklearn.cluster import DBSCAN
```

```python
healthy_male = pd.read_csv('clean_health_male_condition.csv')
healthy_male = healthy_male.loc[:, ~healthy_male.columns.str.contains('^Unnamed')]
healthy_female = pd.read_csv('clean_health_female_condition.csv')
healthy_female = healthy_female.loc[:, ~healthy_female.columns.str.contains('^Unnamed')]
healthy_female = healthy_female.drop(41)
# see check check 
park_male = pd.read_csv('clean_parkinsons_male_condition.csv')
park_male = park_male.loc[:, ~park_male.columns.str.contains('^Unnamed')]
park_female = pd.read_csv('clean_parkinsons_female_condition.csv')
park_female = park_female.loc[:, ~park_female.columns.str.contains('^Unnamed')]
alz_male = pd.read_csv('clean_alzheimers_male_condition.csv')
alz_male = alz_male.loc[:, ~alz_male.columns.str.contains('^Unnamed')]
alz_female = pd.read_csv('clean_alzheimers_female_condition.csv')
alz_female = alz_female.loc[:, ~alz_female.columns.str.contains('^Unnamed')]
sch_male = pd.read_csv('clean_schizophrenia_male_condition.csv')
sch_male = sch_male.loc[:, ~sch_male.columns.str.contains('^Unnamed')]
sch_female = pd.read_csv('clean_schizophrenia_female_condition.csv')
sch_female = sch_female.loc[:, ~sch_female.columns.str.contains('^Unnamed')]
bi_male = pd.read_csv('clean_bipolar_male_condition.csv')
bi_male = bi_male.loc[:, ~bi_male.columns.str.contains('^Unnamed')]
bi_female = pd.read_csv('clean_bipolar_female_condition.csv')
bi_female = bi_female.loc[:, ~bi_female.columns.str.contains('^Unnamed')]
ep_male = pd.read_csv('clean_epilepsy_male_condition.csv')
ep_male = ep_male.loc[:, ~ep_male.columns.str.contains('^Unnamed')]
ep_female = pd.read_csv('clean_epilepsy_female_condition.csv')
ep_female = ep_female.loc[:, ~ep_female.columns.str.contains('^Unnamed')]
dep_male = pd.read_csv('clean_depression_male_condition.csv')
dep_male = dep_male.loc[:, ~dep_male.columns.str.contains('^Unnamed')]
dep_female = pd.read_csv('clean_depression_female_condition.csv')
dep_female = dep_female.loc[:, ~dep_female.columns.str.contains('^Unnamed')]
bi2 = pd.read_csv('more_bipolar.csv')
bi2 = bi2.loc[:, ~bi2.columns.str.contains('^Unnamed')]
dep2 = pd.read_csv('more_dep.csv')
dep2 = dep2.loc[:, ~dep2.columns.str.contains('^Unnamed')]
sch2 = pd.read_csv('more_sch.csv')
sch2 = sch2.loc[:, ~sch2.columns.str.contains('^Unnamed')]

combined_df = pd.concat([healthy_male, healthy_female, park_male, park_female, alz_male, alz_female, 
                        sch_male, sch_female, bi_male, bi_female, ep_male, ep_female, dep_male, dep_female, 
                        bi2, dep2, sch2])
combined_df.fillna(0, inplace=True)
```

```python
neurological_disorders = ['Parkinsons', 'Alzheimers', 'Health', 'Schizophrenia', 'Bipolar Disorder', 
                          'Depression', 'Epilepsy']

combined_df['Condition'] = combined_df['Condition'].apply(lambda x: ', '.join(
    [condition for condition in x.split(', ') if condition in neurological_disorders]) 
                                                          if isinstance(x, str) else x)
```

```python
extra_info = combined_df[['Run ID', 'Sex', 'Age', 'Country', 'BMI', 'Condition']]
label = combined_df['Condition']
label
```

```python
microbiomes = combined_df.drop(['Run ID', 'Sex', 'Age', 'Country', 'BMI', 'Condition'], axis=1)
```

```python
microbiomes
```

```python
rows_sum_to_zero = (microbiomes.sum(axis=1) == 0)
rows_with_zero_sum = microbiomes[rows_sum_to_zero]
rows_with_zero_sum
```

```python
microbiomes
```

```python
row_sums = microbiomes.select_dtypes(include='number').sum(axis=1)

microbiomes_relative_abundance = microbiomes.select_dtypes(include='number').div(row_sums, axis=0)

big_df_ra = pd.concat([extra_info, microbiomes_relative_abundance], axis=1)
```

```python
microbiomes_relative_abundance
```

```python
big_df_ra
big_df_ra.to_csv('combined_data_with_extra_data.csv')
```

```python
kmeans = KMeans(n_clusters=9, random_state=42, n_init=10)
predicted_clusters = kmeans.fit_predict(microbiomes_relative_abundance)

ari3 = adjusted_rand_score(label, predicted_clusters)
print("Adjusted Rand Index (Hier.): ", ari3)
nmi3 = normalized_mutual_info_score(label, predicted_clusters)
print("Normalized Mutual Information (Hier.): ",nmi3)
```
