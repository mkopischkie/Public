```python
import pandas as pd
```

```python
df = pd.read_csv('extra_sch_data.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df
```

```python
df['File Name'] = df['File Name'].str.replace('.fastq', '', regex=False)
df = df.rename(columns={'File Name': 'Run ID'})
df['Condition'] = 'Schizophrenia'
new_order = ['Run ID', 'Condition'] + [col for col in df.columns if col not in ['Run ID', 'Condition']]
df = df[new_order]
```

```python
df.columns
```

```python
df['Uncultured Organism'] += df[['Bacterium', 'Uncultured']].sum(axis=1)
df['Ruminococcus'] += df[['Ruminococcaceae', 'Rumen Bacterium']].sum(axis=1)
df['Clostridium'] += df[['Clostridiales']].sum(axis=1)
```

```python
df = df.drop(['Bacterium', 'Uncultured', 'Ruminococcaceae', 'Rumen Bacterium', 'Clostridiales', 
             'Lachnospiraceae', 'Chain'], axis=1)
```

```python
numeric_cols = df.select_dtypes(include=['number'])

df[numeric_cols.columns] = numeric_cols.div(numeric_cols.sum(axis=1), axis=0)
```

```python
zero_counts = (df == 0.0).sum()
zero_counts
```

```python
threshold = 0.97 * len(df)
filtered = zero_counts[zero_counts <= threshold]
filtered
```

```python
filtered_df = df[filtered.index]
filtered_df.columns
```

```python
filtered_df
```

```python
df1 = pd.read_csv('associated_sch_data.csv')
df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
df1 = df1.rename(columns={'Run': 'Run ID'})
df1
```

```python
merged_df = pd.merge(df1, filtered_df, on='Run ID', how='inner')
merged_df.rename(columns={'sex': 'Sex'}, inplace=True)
merged_df
```

```python
numeric_cols = merged_df.select_dtypes(include=['number'])
rows_sum_to_zero = (numeric_cols.sum(axis=1) == 0)
rows_with_zero_sum = merged_df[rows_sum_to_zero]
rows_with_zero_sum
```

```python
merged_df = merged_df[~rows_sum_to_zero]
```

```python
numeric_cols = merged_df.select_dtypes(include=['number'])
rows_sum_to_zero = (numeric_cols.sum(axis=1) == 0)
rows_with_zero_sum = merged_df[rows_sum_to_zero]
rows_with_zero_sum
```

```python
merged_df.to_csv('more_sch.csv')
```
