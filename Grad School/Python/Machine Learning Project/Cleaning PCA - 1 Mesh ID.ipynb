```python
import pandas as pd
```

## Alzheimers

```python
alzheimers_male = pd.read_csv('full_panel_condition_alz_male.csv')
alzheimers_male = alzheimers_male.loc[:, ~alzheimers_male.columns.str.contains('^Unnamed')]

```

```python
alzheimers_male.head()
```

```python
alzheimers_male['Condition'] = 1
alzheimers_male.rename(columns={'Condition': 'Is Alzheimers?'}, inplace=True)
```

```python
alzheimers_male.head()
```

```python
alzheimers_female = pd.read_csv('full_panel_condition_alz_female.csv')
alzheimers_female = alzheimers_female.loc[:, ~alzheimers_female.columns.str.contains('^Unnamed')]
```

```python
alzheimers_female.head()
```

```python
alzheimers_female['Condition'] = 1
alzheimers_female.rename(columns={'Condition': 'Is Alzheimers?'}, inplace=True)
```

```python
alzheimers_female.head()
```

```python
healthy_male = pd.read_csv('full_panel_condition_healthy_male.csv')
healthy_male = healthy_male.loc[:, ~healthy_male.columns.str.contains('^Unnamed')]
```

```python
healthy_male.head()
```

```python
healthy_male['Condition'] = 0
healthy_male.rename(columns={'Condition': 'Is Alzheimers?'}, inplace=True)
```

```python
healthy_male.head()
```

```python
healthy_female = pd.read_csv('full_panel_condition_healthy_female.csv')
healthy_female = healthy_female.loc[:, ~healthy_female.columns.str.contains('^Unnamed')]
```

```python
healthy_female.head()
```

```python
healthy_female['Condition'] = 0
healthy_female.rename(columns={'Condition': 'Is Alzheimers?'}, inplace=True)
```

```python
healthy_female.head()
```

```python
is_alz_df = pd.concat([alzheimers_male, alzheimers_female, healthy_male, healthy_female])
```

```python
is_alz_df = is_alz_df.fillna(0)
```

```python
is_alz_df.head()
```

```python
is_alz_df.to_csv('alzheimers_healthy_for_pca_one_condition.csv')
```
