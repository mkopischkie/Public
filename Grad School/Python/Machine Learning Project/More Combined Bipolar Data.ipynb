```python
import pandas as pd
```

```python
df = pd.read_csv('extra_bipolar_data.csv')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.fillna(0, inplace=True)
df.head()
```

```python
df['File Name'] = df['File Name'].str.replace('.fastq', '', regex=False)
df = df.rename(columns={'File Name': 'Run ID'})
```

```python
df.columns
```

```python
df['Uncultured Organism'] += df[['Bacterium', 'Uncultured', 'Unidentified']].sum(axis=1)
df['Bacteroides'] += df[['Bacteroidales', 'Bacteroidetes', 'Bacteroidaceae']].sum(axis=1)
df['Oscillibacter'] += df[['Oscillospiraceae']].sum(axis=1)
df['Blautia'] += df['Lachnospiraceae'] / 2
df['Roseburia'] += df['Lachnospiraceae'] / 2
```

```python
df = df.drop(['Bacterium', 'Uncultured', 'Unidentified', 'Control', 'Moo', 'Evans', 'Mangal', 'Mangala', 
              'Marseille', 'Bacteroidales', 'Bacteroidetes', 'Bacteroidaceae', 'Oscillospiraceae', 
              'Lachnospiraceae'], axis=1)
```

```python
df.columns
```

```python
numeric_cols = df.select_dtypes(include=['number'])

df[numeric_cols.columns] = numeric_cols.div(numeric_cols.sum(axis=1), axis=0)
```

```python
df.columns
```

```python
zero_counts = (df == 0.0).sum()
zero_counts
```

```python
threshold = 0.97 * len(df)
filtered = zero_counts[zero_counts <= threshold]
filtered
```

```python
filtered_df = df[filtered.index]
filtered_df.columns
```

```python
filtered_df = filtered_df.rename(columns={'File Name': 'Run ID'})
```

```python
df1 = pd.read_csv('associated_bi_data.csv')
df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
```

```python
df1.rename(columns={'Run': 'Run ID'}, inplace=True)
```

```python
df1
```

```python
df1.columns
filtered_df.columns
```

```python
merged_df = pd.merge(df1, filtered_df, on='Run ID', how='inner')
merged_df
```

```python
merged_df = merged_df.rename(columns={'Host_disease': 'Condition', 'geo_loc_name_country': 'Country', 
                                     'Host_age': 'Age', 'host_body_mass_index': 'BMI', 'host_sex': 'Sex'})
merged_df = merged_df.drop(['ETHNICITY'], axis=1)

dictionary = {'Bipolar Disorder': 'Bipolar Disorder',
              'Healthy Control': 'Health'
}

merged_df['Condition'] = merged_df['Condition'].replace(dictionary)
```

```python
merged_df
```

```python
merged_df['Condition'].unique()
```

```python
numeric_cols = merged_df.select_dtypes(include=['number'])
rows_sum_to_zero = (numeric_cols.sum(axis=1) == 0)
rows_with_zero_sum = merged_df[rows_sum_to_zero]
rows_with_zero_sum
```

```python
merged_df = merged_df[~rows_sum_to_zero]
```

```python
numeric_cols = merged_df.select_dtypes(include=['number'])
rows_sum_to_zero = (numeric_cols.sum(axis=1) == 0)
rows_with_zero_sum = merged_df[rows_sum_to_zero]
rows_with_zero_sum
```
