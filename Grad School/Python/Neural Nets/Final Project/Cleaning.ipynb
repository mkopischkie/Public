```python
df['Hazard_clean'] = df['Hazard'].str.replace(r'^.*?:\s*', '', regex=True)  # remove everything before the colon
df['Hazard_clean'] = df['Hazard_clean'].str.replace(r'\[.*?\]', '', regex=True)  # remove brackets and content
df['Hazard_clean'] = df['Hazard_clean'].str.strip()  # remove leading/trailing spaces
```

```python
hazards_to_remove = [
    'Aggregated GHS information provided per 6 reports by companies from 1 notifications to the ECHA C&L Inventory.',
    'Reported as not meeting GHS hazard criteria per 6 of 6 reports by companies.',
    'There are 0 notifications provided by 0 of 6 reports by companies with hazard statement code(s).',
    'Information may vary between notifications depending on impurities, additives, and other factors. The percentage value in parenthesis indicates the notified classification ratio from companies that provide hazard codes. Only hazard codes with percentage values above 10% are shown. For more detailed information, please visit  ECHA C&L website.',
    'Aggregated GHS information provided per 2465 reports by companies from 14 notifications to the ECHA C&L Inventory.',
    'Reported as not meeting GHS hazard criteria per 2223 of 2465 reports by companies.',
    'There are 12 notifications provided by 242 of 2465 reports by companies with hazard statement code(s).',
    'Aggregated GHS information provided per 58 reports by companies from 4 notifications to the ECHA C&L Inventory.',
    'Reported as not meeting GHS hazard criteria per 55 of 58 reports by companies.',
    'There are 3 notifications provided by 3 of 58 reports by companies with hazard statement code(s).',
    'Aggregated GHS information provided per 41 reports by companies from 2 notifications to the ECHA C&L Inventory.',
    'Aggregated GHS information provided per 42 reports by companies from 2 notifications to the ECHA C&L Inventory.',
    'Aggregated GHS information provided per 2 reports by companies from 2 notifications to the ECHA C&L Inventory.',
    'Reported as not meeting GHS hazard criteria per 2 of 2 reports by companies.',
    'There are 0 notifications provided by 0 of 2 reports by companies with hazard statement code(s).',
    'Warning', 'Danger'
]

df = df[~df['Hazard_clean'].isin(hazards_to_remove)]
```

```python
df = df.dropna(subset=['Hazard_clean'])
```

```python
df
```

```python
df['Hazard_clean'].unique().tolist()
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('unborn child|fertility|children', case=False, na=False),
    'May cause danger to unborn child, breastfed child, or fertility'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('aquatic life', case=False, na=False),
    'Toxic to aquatic life'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('organs', case=False, na=False),
    'May cause damage to organs'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('cancer', case=False, na=False),
    'May cause cancer'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('genetic defects', case=False, na=False),
    'May cause genetic defects'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('eye', case=False, na=False),
    'Causes eye damage or irritation'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    df['Hazard_clean'].str.contains('respiratory|breathing', case=False, na=False),
    'May cause breathing difficulties, respiratory irritation, or asthma symptoms'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    (df['Hazard_clean'] == 'Harmful in contact with skin') | 
    (df['Hazard_clean'] == 'May cause an allergic skin reaction') | 
    (df['Hazard_clean'] == 'Toxic in contact with skin') | 
    (df['Hazard_clean'] == 'Causes severe skin burns and eye damage') | 
    (df['Hazard_clean'] == 'Toxic if swallowed or in contact with skin') |
    (df['Hazard_clean'] == 'Harmful if swallowed, in contact with skin or if inhaled') |
    (df['Hazard_clean'] == 'Causes skin irritation'),
    'May cause skin irritation, damage, or burns'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    (df['Hazard_clean'] == 'Harmful if swallowed') | 
    (df['Hazard_clean'] == 'Harmful if inhaled'),
    'Harmful if swallowed or inhaled'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    (df['Hazard_clean'] == 'Fatal if swallowed') | 
    (df['Hazard_clean'] == 'Fatal if swallowed or if inhaled') |
    (df['Hazard_clean'] == 'Fatal in contact with skin') |
    (df['Hazard_clean'] == 'Fatal if inhaled'),
    'Fatal if swallowed, inhaled, or in contact with skin'
)
```

```python
df['Hazard_clean'] = df['Hazard_clean'].mask(
    (df['Hazard_clean'] == 'Toxic if swallowed') | 
    (df['Hazard_clean'] == 'Toxic if inhaled'),
    'Toxic if swallowed or inhaled'
)
```
