```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm, ticker
import matplotlib.colors as colors
```

```python
h = 6.626*10**(-34)
c = 3.0*10**8
k = 1.38*10**(-23)

def planck(wavelength,temp):
    num = (8 * np.pi * h * c**2)
    denom = (wavelength**5 * (np.e**((h * c)/(wavelength * k * temp))-1))
    return(num/denom)
```

```python
print(planck(100e-9,5800))
```

```python
#range of wavelengths and plot the blackbody curve 
wav = np.arange(100e-9,2e-6,1e-9)

intensity5800 = planck(wav,5800)
print(intensity5800)
intensity4500 = planck(wav,4500)
intensity7000 = planck(wav,7000)
```

```python
print(wav, intensity5800)
```

```python


plt.plot(wav, intensity5800*1e-14, c='m')
plt.plot(wav, intensity4500*1e-14, c='g')
plt.plot(wav, intensity7000*1e-14, c='c')
plt.xlabel('Wavelength (m)')
plt.ylabel('Intensity (Wm^-3)')
plt.title('Blackbody Radiation')
plt.show()
```

```python
x = wav
y = intensity5800*1e-14
y2 = intensity4500*1e-14
y3 = intensity7000*1e-14

ymax = max(intensity5800)
print(ymax)
xmax = wav[np.argmax(intensity5800)]
print(xmax)
y2max = max(intensity4500)
print(y2max)
x2max = wav[np.argmax(intensity4500)]
print(x2max)
y3max = max(intensity7000)
print(y3max)
x3max = wav[np.argmax(intensity7000)]
print(x3max)
```

```python
fig = plt.figure()
plt.plot(x, y, c='m')
plt.plot(x, y2, c='g')
plt.plot(x, y3, c='c')
plt.xlabel('Wavelength (m)')
plt.ylabel('Intensity*1e-14')
plt.title('Blackbody Radiation')

plt.annotate("x={:.1e}, y={:.2e}".format(xmax, ymax), xy=(xmax, ymax*1e-14), xytext=(0.85*1e-6, ymax*1e-14+0.75), arrowprops=dict(facecolor='magenta'),)
plt.annotate("x={:.1e}, y={:.2e}".format(x2max, y2max), xy=(x2max, y2max*1e-14), xytext=(1.25*1e-6, y2max*1e-14+0.75), arrowprops=dict(facecolor='green'),)
plt.annotate("x={:.1e}, y={:.2e}".format(x3max, y3max), xy=(x3max, y3max*1e-14), xytext=(0.5*1e-6, y3max*1e-14+0.75), arrowprops=dict(facecolor='cyan'),)

plt.ylim(-0.1,10.0)
plt.show()
fig.savefig('graph.png')
```

```python
fig = plt.figure()
fig.set_figwidth(8)
fig.set_figheight(8)
ax = plt.axes(projection='3d')

wave = np.linspace(100e-9,2e-6, num=1000)
temperature = np.linspace(4500,7000, num=1000)
wave,temperature = np.meshgrid(wave,temperature)
inten = planck(wave, temperature)

plt.xlabel('Wavelength (m)', fontsize = 10)
plt.ylabel('Temperature (K)', fontsize = 10)
ax.set_zlabel('Intensity', fontsize = 10)
plt.title('Blackbody Radiation')

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)

ax.plot_surface(wave,temperature,inten,norm=colors.LogNorm(vmin=inten.min(), vmax=inten.max()),cmap=cm.copper)
#ax.plot_surface(wave,temperature,inten,cmap=inten,norm=colors.LogNorm())
#ax.pcolormesh(wave, temperature, inten, vmin=-1., vmax=1., cmap='copper')
plt.show()
#fig.savefig('3dgraph.png')
```
